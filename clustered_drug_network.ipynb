```python
drug_per_doc = []
for x in drug_df.columns:
#     if x
    mentioned_drug = [x for x in drug_df[x].values if x!='']
#     print(mentioned_drug)
    drug_per_doc.append(mentioned_drug)
```

```python
all_drugs = []
for x in drug_per_doc:
    all_drugs.extend(x)
all_drugs = list(set(all_drugs))
```

```python
len(all_drugs)
```

```python
all_drugs
```

```python
connected_dic = {}
for drug in all_drugs:
    drug_contain_idx =[]
    for idx in range(len(drug_per_doc)):
        if drug in drug_per_doc[idx]:
            drug_contain_idx.append(idx+1)
    print(drug_contain_idx)
    if len(drug_contain_idx)==1:
        connected_dic[drug] = [(drug_contain_idx[0],drug_contain_idx[0] )]
        continue
    connected_node = list(combinations(drug_contain_idx, 2))
    connected_dic[drug] = connected_node
#     print(connected_node)
```

```python
for x,y in connected_dic.items():
#     print(y)
    for z in y:
        c =list(z)
        c.append(x)
        print(str(c[0])+'\t'+str(c[1])+'\t'+str(c[2]))
```
